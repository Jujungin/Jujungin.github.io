# DSA with python

## 1. Min, Max

```python
myArray = [10, 9, 8, 7, 3, 6, 2, 1, 5, 4, 4, 0, 200]
```

```python
minVal = myArray[0]
maxVal = myArray[0]
```

```python
for i in myArray:
  if i < minVal:
    minVal = i

for i in myArray:
  if i > maxVal:
    maxVal = i
```

```python
minVal, min(myArray), maxVal, max(myArray)
```

## 2. 피보나치 수를 구해보장

```python
defaultVal = 10
answer = [0, 1]
```

```python
for i in range(2, defaultVal):
  answer.append(answer[i-1] + answer[i-2])
```

```python
answer
```

### 나는 무한대로 가고싶고 원하는 값을 넘어가면 중단하겠다!

```python
separateVal = 100
answer = [0, 1]
```

```python
while True:
  nxt = answer[-1] + answer[-2]
  if nxt > separateVal:
    break
  answer.append(nxt)
```

#### 컴포넌트를 분리시켜 볼깡?

```python
def fibo(answer, separateVal):
  while True:
    nxt = answer[-1] + answer[-2]
    if nxt > separateVal:
      break
    answer.append(nxt)

  return answer
```

```python
answer = [0, 1]
separateVal = 100
```

```python
fibo(answer, separateVal)
```

#### Recursive로는?

```python
def Fibo(answer, separateVal):
  nxt = answer[-1] + answer[-2]
  if nxt > separateVal:
    return answer
  else:
    answer.append(nxt)
    return Fibo(answer, separateVal)
```

```python
Fibo(answer, separateVal)
```
